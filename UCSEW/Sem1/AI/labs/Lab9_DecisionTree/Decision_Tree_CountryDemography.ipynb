```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
country_data = pd.read_csv('country-demographic-data.csv')
country_data
```

```python
# Birthrate and Internet Users as inputs
x = country_data.iloc[:, 2:4]
x
```

```python
#Income group as Output
y = country_data.iloc[:, -1]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
xtrain.shape, xtest.shape
```

```python
ytrain.shape, ytest.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='gini',
                                   random_state=100,
                                   max_depth=4, 
                                   min_samples_leaf=3)
classifier.fit(xtrain, ytrain)
```

```python
y_pred = classifier.predict(xtest)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(ytest,y_pred)
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True)

# Multi class confusion matrix (4x4)
# 4 Predicted field values * 4 Actual field values
```

```python
accuracy_score(ytest, y_pred)
```

```python
print(classification_report(ytest, y_pred))
```

```python
plt.figure(figsize=(25,20))
from sklearn import tree
tree.plot_tree(classifier, 
               feature_names = x.columns,
               class_names= y.unique(),
               filled=True)
```
