
#  Identifiers for Variables
An Identifier is a name used to identify a variable, function, class, module or object.
     An identifier must start with an alphabet (A..Z or a..z) or underscore ( _ ).
     Identifiers may contain digits (0 .. 9)
     Python identifiers are case sensitive i.e. uppercase and lowercase letters are distinct.
     Identifiers must not be a python keyword.
     Python does not allow punctuation character such as %,$, @ etc., within identifiers.

# Python Data Types

Data types are the classification or categorization of data items. It represents the kind of value that tells what operations can be performed on a particular data. Since everything is an object in Python programming, data types are actually classes and variables are instance (object) of these classes.
  
   
    

```python
#Integers
#we can represent using decimal,binary,octal,hexadecimal
a=213 #decimal representation
b=0b111 #binary representation
c=0o347 #octal representation
d=0xF #hexadecimal representation
print(a,b,c,d)

#Float
f1=10.3
f2=10E4   #10*10pow(4)
f3=10E-4  #10*10pow(-4)
print(f1,f2,f3)

#Complex
c1=10+3j
c2=2+4j
cn=c1+c2
print(cn)

#To check type of all above variables:
print(type(a),type(b),type(c),type(d),type(f1),type(f2),type(f3),type(c1),type(c2))

#Boolean   True and False
x=True
y=False
print(type(x),type(y))

#Non zero numbers are treated as True
xx=bool(3)  #when you try to cast to boolean type only 0 will be treated as False
yy=bool(0)
zz=bool(.5)
print(xx,yy,zz)
ss=bool("") #empty string is treated as false
se=bool("sdsdd")
print(ss,se)
```
