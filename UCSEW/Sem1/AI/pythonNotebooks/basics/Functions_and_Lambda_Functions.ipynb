# Functions in  python


```python
#define function
def summ():
    return "add",10
def diffe():
    return "sub",20
def mul():
    return "miul",30


x,y=summ()
print(x,y)
print(summ())
print(diffe())
print(summ())
#print(mul(10,20))
```

```python
#call function
def x(x):
    print("hai")
    print(x)
x(2)
```

```python
#required arguments
def x(a,b):
    print("hai")
x("hello","hai")
```

```python
#keyword Arguments
def x(y,name):
    print("hello")
x(10,name="hello")
```

```python
#Default Arguments
def x(a,b=100):
    print(a,b)
x(10,200)
```

```python
#Variable length arguments
def dis(y,*x):
    print(x,y)
    for i in x:
        print(i)
dis([1,2,3],[3,4,5],6)    


```

```python
#Example program Functions
def calc(marks):
    return max(marks), min(marks)

def sumAll(*x):
    return sum(x)


marks=[]
loop = int(input("Enter number of students:"))
for i in range (loop):
    mark = int(input("Enter mark:"))
    marks.append(mark)
maxmark,minmark = calc(marks)
print(marks)
print("max marks from the student is: ", maxmark)
print("Min marks from the student is: ", minmark)
    
```

```python
#Deleting an item in List
list1 = [2.5,6,8.7,9,11,12,15,10]
#pop to exclude
print(list1.pop(1))
del list1[1]
print(list1)
list1.remove(2.5)
print(list1)
```

```python
# Write a function 
def add10(x):
    x = x + 10
    return x
```

```python
# Lambda Function - Small anonymous function
# Can take any number of arguments, but can have only one expression
# Syntax - lambda arguments: expression
x = lambda a: a+10
x(15)
```
